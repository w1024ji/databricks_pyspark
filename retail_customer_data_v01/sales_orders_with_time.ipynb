```python
table_name = "databricks_simulated_retail_customer_data.v01.sales_orders"

orders_df = spark.read.table(table_name)

display(orders_df.limit(5))
```

```python
orders_df.printSchema()
```

```python
from pyspark.sql.functions import col, from_unixtime

transformed_df = orders_df.withColumn("order_timestamp", from_unixtime(col("order_datetime")).cast("timestamp")) \
                          .withColumn("customer_name_upper", col("customer_name"))

display(transformed_df.select("customer_name_upper", "order_datetime", "order_timestamp").limit(10))
```

##### what time does ppl buy the most?

```python

from pyspark.sql.functions import hour

df_with_hour = transformed_df.withColumn("order_hour", hour(col("order_timestamp")))

hourly_trend = df_with_hour.groupBy("order_hour").count().orderBy("order_hour")

clean_hourly_trend = hourly_trend.filter(col("order_hour").isNotNull())

display(clean_hourly_trend)
```
